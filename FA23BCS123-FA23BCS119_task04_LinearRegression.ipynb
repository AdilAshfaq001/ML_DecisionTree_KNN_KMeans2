Task 1: Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdilAshfaq001/ML_DecisionTree_KNN_KMeans2/refs/heads/master/canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt  # Make sure this import is present

plt.xlabel('year')  # Use plt.xlabel instead of plot.xlabel
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'])
```

```python
reg = linear_model.LinearRegression()
reg.fit(df['year'].values.reshape(-1, 1), df['per capita income (US$)'].values)
```

```python
# Reshape the input to a 2D array
prediction = reg.predict(np.array([[2020]]))
print(prediction)
```

Task 2: Multiple Linear Regression

```python
import numpy as np # linear algebra
import pandas as pd 
from sklearn import linear_model
```

```python
df =pd.read_csv("https://raw.githubusercontent.com/AdilAshfaq001/ML_DecisionTree_KNN_KMeans2/refs/heads/master/hiring.csv")
df.head()
```

```python
df.experience = df.experience.fillna(0)
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(8.0)
df
```

```python
# Assuming df is your DataFrame
df = df.replace({
    'five': 5,
    'two': 2,
    'seven': 7,
    'three': 3,
    'ten': 10,
    'eleven': 11
})

# Explicitly infer datatypes (removes the FutureWarning about downcasting)
df = df.infer_objects(copy=False)
```

```python
df.experience
```

```python
df['experience'] = pd.to_numeric(df['experience'])
df
```

```python
df['experience'] = df['experience'].fillna(0)
df
```

```python
X = df.drop('salary($)', axis='columns').values
y = df['salary($)'].values

model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
model.predict([[2,9,6]])
```

```python
model.predict([[12,10,10]])
```
